```python
# Import required libraries
import tkinter as tk

# Function to update the input field whenever a button is pressed
def btn_click(item):
    current = input_text.get()
    input_text.set(current + str(item))

# Function to clear the input field
def clear():
    input_text.set("")

# Function to evaluate the expression entered in the input field
def evaluate():
    try:
        result = str(eval(input_text.get()))
        input_text.set(result)
    except:
        input_text.set("Error")

# Create main window object
root = tk.Tk()
root.title("Simple Calculator")

# Define the input field using StringVar to update dynamically
input_text = tk.StringVar()

# Create the input field (entry widget)
input_frame = tk.Frame(root)
input_frame.pack()

input_field = tk.Entry(input_frame, textvariable=input_text, width=25, font=('arial', 18, 'bold'), bd=5)
input_field.grid(row=0, column=0)
input_field.pack(ipady=10)

# Create a frame for the buttons
btns_frame = tk.Frame(root)
btns_frame.pack()

# Define buttons
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('C', 4, 1), ('=', 4, 2), ('+', 4, 3)
]

# Loop to create buttons dynamically
for (text, row, col) in buttons:
    if text == '=':
        btn = tk.Button(btns_frame, text=text, width=10, height=3, command=evaluate)
    elif text == 'C':
        btn = tk.Button(btns_frame, text=text, width=10, height=3, command=clear)
    else:
        btn = tk.Button(btns_frame, text=text, width=10, height=3, command=lambda t=text: btn_click(t))
    btn.grid(row=row, column=col)

# Start the GUI event loop
root.mainloop()
```

```python
import tkinter as tk
import random

class SnakeGame(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Snake Game")
        self.geometry("400x400")
        self.resizable(False, False)

        # Game variables
        self.snake = [(100, 100), (90, 100), (80, 100)]
        self.food_position = self.get_food_position()
        self.direction = "Right"
        self.running = True
        self.score = 0

        # Create canvas for drawing the game
        self.canvas = tk.Canvas(self, bg="black", height=400, width=400)
        self.canvas.pack()

        # Bind keys to control the snake
        self.bind("<Up>", self.go_up)
        self.bind("<Down>", self.go_down)
        self.bind("<Left>", self.go_left)
        self.bind("<Right>", self.go_right)

        # Start game loop
        self.update_game()

    # Create random position for food
    def get_food_position(self):
        return (random.randint(1, 39) * 10, random.randint(1, 39) * 10)

    # Move the snake
    def move_snake(self):
        head_x, head_y = self.snake[0]

        if self.direction == "Up":
            new_head = (head_x, head_y - 10)
        elif self.direction == "Down":
            new_head = (head_x, head_y + 10)
        elif self.direction == "Left":
            new_head = (head_x - 10, head_y)
        elif self.direction == "Right":
            new_head = (head_x + 10, head_y)

        # Add new head to snake
        self.snake = [new_head] + self.snake

        # Check if snake has eaten food
        if new_head == self.food_position:
            self.food_position = self.get_food_position()  # Generate new food
            self.score += 1  # Increase score
        else:
            self.snake.pop()  # Remove the tail unless the snake has eaten food

    # Check if snake has collided with itself or the wall
    def check_collisions(self):
        head_x, head_y = self.snake[0]

        # Wall collision
        if head_x < 0 or head_x >= 400 or head_y < 0 or head_y >= 400:
            self.running = False

        # Self collision
        if len(self.snake) != len(set(self.snake)):
            self.running = False

    # Game over function
    def game_over(self):
        self.canvas.create_text(200, 200, fill="white", font="Arial 20", text=f"Game Over! Score: {self.score}")
    
    # Update the game loop
    def update_game(self):
        if self.running:
            self.move_snake()
            self.check_collisions()

            # Redraw everything on the canvas
            self.canvas.delete(tk.ALL)
            self.draw_snake()
            self.draw_food()

            # Continue game loop
            self.after(100, self.update_game)
        else:
            self.game_over()

    # Draw the snake
    def draw_snake(self):
        for segment in self.snake:
            self.canvas.create_rectangle(segment[0], segment[1], segment[0] + 10, segment[1] + 10, fill="green")

    # Draw the food
    def draw_food(self):
        self.canvas.create_rectangle(self.food_position[0], self.food_position[1], 
                                     self.food_position[0] + 10, self.food_position[1] + 10, fill="red")

    # Controls for the snake
    def go_up(self, event):
        if self.direction != "Down":
            self.direction = "Up"

    def go_down(self, event):
        if self.direction != "Up":
            self.direction = "Down"

    def go_left(self, event):
        if self.direction != "Right":
            self.direction = "Left"

    def go_right(self, event):
        if self.direction != "Left":
            self.direction = "Right"

if __name__ == "__main__":
    game = SnakeGame()
    game.mainloop()
```
